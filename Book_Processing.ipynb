#### Clean the JSON file as well 

```python
import json

def fix_splitted_words_in_json(data):
    if isinstance(data, dict):
        return {key: fix_splitted_words_in_json(value) for key, value in data.items()}
    elif isinstance(data, list):
        return [fix_splitted_words_in_json(element) for element in data]
    elif isinstance(data, str):
        return fix_splitted_words_in_text(data)
    else:
        return data

# Read the input JSON file, fix split words, and save the corrected JSON back to the file
def process_json_file(input_file_path, output_file_path):
    with open(input_file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    
    # Apply the word-fixing function to the JSON data
    fixed_data = fix_splitted_words_in_json(data)
    
    # Write the corrected JSON data to the output file
    with open(output_file_path, 'w', encoding='utf-8') as file:
        json.dump(fixed_data, file, ensure_ascii=False, indent=4)

# Example usage
input_file = 'TOC.json'
output_file = 'TOC_clean.json'
process_json_file(input_file, output_file)
```

```python
import json
import pandas as pd

def flatten_toc(node, parent_info=None, sections_list=None):
    if sections_list is None:
        sections_list = []
    if parent_info is None:
        parent_info = {'chapter_number': '', 'chapter_title': ''}

    # Safe check for 'type' key
    node_type = node.get('type')
    if node_type == 'Chapter':
        # It's a chapter
        chapter_number = node.get('number', '')
        chapter_title = node.get('title', '')
        # Update parent_info
        parent_info.update({
            'chapter_number': chapter_number,
            'chapter_title': chapter_title
        })
        # Process children
        for child in node.get('children', []):
            flatten_toc(child, parent_info.copy(), sections_list)
    else:
        # It's a section or subsection
        section_number = node.get('number', '')
        section_title = node.get('title', '')
        page = node.get('page', None)
        # Add this section to the list
        sections_list.append({
            'chapter_number': parent_info['chapter_number'],
            'chapter_title': parent_info['chapter_title'],
            'section_number': section_number,
            'section_title': section_title,
            'page': page
        })
        # Process any children
        for child in node.get('children', []):
            flatten_toc(child, parent_info.copy(), sections_list)
    return sections_list

# Read the TOC JSON file
with open('TOC_clean.json', 'r', encoding='utf-8') as f:
    toc_json_list = json.load(f)

# Flatten the TOC into a list of sections
sections_list = []
for chapter in toc_json_list:
    flatten_toc(chapter, sections_list=sections_list)

# Optional: Convert to DataFrame
df_sections = pd.DataFrame(sections_list)
```

```python
# Read the extracted text and process it
pages_text_list = []
current_page_num = None
current_page_text = ''

with open('data/risk_management_clean.txt', 'r', encoding='utf-8') as f:
    for line in f:
        if line.startswith('--- Page '):
            # Save the previous page
            if current_page_num is not None:
                pages_text_list.append((current_page_num, current_page_text))
            # Start a new page
            current_page_num = int(line.strip().split(' ')[2])
            current_page_text = ''
        else:
            current_page_text += line

    # Save the last page
    if current_page_num is not None:
        pages_text_list.append((current_page_num, current_page_text))

# Create a list of lines with page numbers
lines = []
for page_num, page_text in pages_text_list:
    line_texts = page_text.splitlines()
    for line_num, line_text in enumerate(line_texts):
        lines.append((page_num, line_text.strip()))
```

```python
import re

def find_section_start_idx(section_number, section_title, lines, start_page_num=None):
    # Prepare the search pattern
    stop_words = set(['the', 'and', 'of', 'in', 'to', 'for', 'with', 'on', 'by', 'is', 'are', 'a', 'an'])
    title_words = [word for word in re.findall(r'\w+', section_title.lower()) if word not in stop_words]
    if not title_words:
        # If no significant words, just search for section number
        pattern = re.escape(section_number)
    else:
        # Build a pattern that matches the section number and some words from the title
        pattern = re.escape(section_number) + r'.*' + r'.*'.join(map(re.escape, title_words[:2]))
    # Use case-insensitive search
    pattern = '(?i)' + pattern
    # Start searching from the line where the page number >= start_page_num
    for idx, (page_num, line_text) in enumerate(lines):
        if start_page_num is not None and page_num < start_page_num:
            continue
        if re.search(pattern, line_text):
            return idx
    return None

# Find the start index for each section
for section in sections_list:
    section_number = section['section_number']
    section_title = section['section_title']
    start_page_num = section.get('page', None)
    start_idx = find_section_start_idx(section_number, section_title, lines, start_page_num)
    if start_idx is not None:
        section['start_idx'] = start_idx
    else:
        print(f"Section {section_number} not found in text.")
        section['start_idx'] = None
```

```python
# Filter sections with valid start indices
sections_with_start_idx = [s for s in sections_list if s['start_idx'] is not None]

# Sort sections by their start index
sections_with_start_idx.sort(key=lambda s: s['start_idx'])

# Collect text for each section
for i, section in enumerate(sections_with_start_idx):
    start_idx = section['start_idx']
    if i + 1 < len(sections_with_start_idx):
        end_idx = sections_with_start_idx[i + 1]['start_idx']
    else:
        end_idx = len(lines)
    # Collect the lines from start_idx to end_idx
    section_lines = lines[start_idx:end_idx]
    section_text = '\n'.join(line_text for page_num, line_text in section_lines)
    section['text'] = section_text
```

```python
import pandas as pd
import re
from collections import Counter

special_characters_list = []

# Iterate through each text entry to find special characters
def find_special_characters(text):
    # Find all non-alphanumeric and non-space characters using regex
    return re.findall(r'[^a-zA-Z0-9\s]', text)

for text in df['text']:
    # Include newline (\n) and tab (\t) explicitly
    if "\n" in text:
        special_characters_list.append("\n")
    if "\t" in text:
        special_characters_list.append("\t")
    special_characters_list.extend(find_special_characters(text))

# Convert list to a dictionary with counts of each special character
special_characters_count = dict(Counter(special_characters_list))

# Print the count of each special character
print("Count of special characters in the 'text' column:", special_characters_count)
```

Delete the special characters

```python
def remove_special_characters(text):
    return re.sub(r'[^a-zA-Z0-9\.\s]', ' ', text)

# Apply the function to the 'text' column
df['text'] = df['text'].apply(remove_special_characters)
```

Delete the \n from the text column

```python
def remove_newline(text):
    return text.replace("\n", " ")

# Apply the function to the 'text' column
df['text'] = df['text'].apply(remove_newline)
```

trim space

```python
# Function to replace multiple spaces with a single space
def trim_spaces(text):
    return re.sub(r'\s+', ' ', text)

# Apply the function to the 'text' column
df['text'] = df['text'].apply(trim_spaces)
```

```python
def prioritize_edges(df: pd.DataFrame) -> pd.DataFrame:
    def prioritize(edge_str):
        edges = edge_str.split(',')
        # Remove 'contextual_proximity' and prioritize the others
        edges = [edge for edge in edges if edge != 'contextual_proximity']
        return edges[0] if edges else 'contextual_proximity'

    df['edge'] = df['edge'].apply(prioritize)
    return df
```

```python
def contextual_proximity(df: pd.DataFrame) -> pd.DataFrame:
    dfg_long = pd.melt(
        df, id_vars=["chunk_id"], value_vars=["node_1","node_2"], value_name="node"
    )

    dfg_long.drop(columns = ["variable"], inplace=True)
    dfg_wide = pd.merge(dfg_long, dfg_long, on="chunk_id", suffixes=("_1", "_2"))
    self_loops_drop  = dfg_wide[dfg_wide["node_1"]==dfg_wide["node_2"]].index
    dfg2 = dfg_wide.drop(index = self_loops_drop).reset_index(drop=True)

    dfg2=(
        dfg2.groupby(["node_1", "node_2"])
        .agg({"chunk_id":[",".join,"count"]})
        .reset_index()
    )

    dfg2.columns = ["node_1", "node_2", "chunk_id", "count"]
    dfg2.replace("", np.nan, inplace=True)
    dfg2.dropna(subset=["node_1", "node_2"], inplace=True)

    dfg2 = dfg2[dfg2["count"] != 1]
    dfg2["edge"] = "contextual_proximity"
    return dfg2

dfg1 = prioritize_edges(dfg1)

dfg2 = contextual_proximity(dfg1)
dfg2.tail()
```

```python
import networkx as nx
G = nx.Graph()

for node in nodes:
    G.add_node(
        str(node)
    )

for index, row in dfg_filtered.iterrows():
    G.add_edge(
        str(row["node_1"]),
        str(row["node_2"]),
        title=row["edge"],
        weight=row['count']
    )
```

```python
communities_generator = nx.community.girvan_newman(G)
top_level_communities = next(communities_generator)
next_level_communities = next(communities_generator)
communities = sorted(map(sorted, next_level_communities))

print("Number of Communities:", len(communities))
print(communities)
```

```python
import seaborn as sns
palette = "hls"

def colors2Community(communities) -> pd.DataFrame:

    p=sns.color_palette(palette, len(communities)).as_hex()
    random.shuffle(p)
    rows = []
    group=0

    for community in communities:
        color = p.pop()
        group+=1
        
        for node in community:
            rows += [{"node": node, "color": color, "group": group}]
    
    df_colors = pd.DataFrame(rows)
    return df_colors
```

```python
# Identify nodes in small communities (5 or fewer nodes)
small_community_nodes = [node for community in communities if len(community) <= 20 for node in community]

# Remove those nodes from the graph
G.remove_nodes_from(small_community_nodes)

# Now reassign colors and groups for the remaining nodes
filtered_communities = [community for community in communities if len(community) > 20]
print("Number of Communities after filtering:", len(filtered_communities))
print("New number of nodes: ", G.number_of_nodes())
```

```python
colors = colors2Community(filtered_communities)
colors.head(5)
```

```python
for index, row in colors.iterrows():
    G.nodes[row['node']]['group'] = row['group']
    G.nodes[row['node']]['color'] = row['color']
    G.nodes[row['node']]['size'] = G.degree[row['node']]
```

```python
import torch
import torch.nn.functional as F
import torch.nn as nn
import torch.optim as optim
from torch_geometric.nn import GATConv, BatchNorm, LayerNorm, TopKPooling
from torch_geometric.data import Data
from torch_geometric.datasets import Planetoid
from sklearn.metrics import f1_score, classification_report
```

```python
# Step 1: Load a Larger Dataset (Cora)
dataset = Planetoid(root='/tmp/Cora', name='Cora')
data = dataset[0]
```

```python
# Step 2: Define the Enhanced GAT Model
class EnhancedGATWithAttention(nn.Module):
    def __init__(self, num_node_features, hidden_channels, num_classes):
        super(EnhancedGATWithAttention, self).__init__()
        
        # Initial GATConv Layers
        self.conv1 = GATConv(num_node_features, hidden_channels, heads=4, concat=True)
        self.norm1 = LayerNorm(hidden_channels * 4)
        
        self.conv2 = GATConv(hidden_channels * 4, hidden_channels, heads=4, concat=True)
        self.norm2 = LayerNorm(hidden_channels * 4)

        self.conv3 = GATConv(hidden_channels * 4, hidden_channels, heads=4, concat=True)
        self.norm3 = LayerNorm(hidden_channels * 4)

        self.conv4 = GATConv(hidden_channels * 4, num_classes, heads=1, concat=False)

        # Dropout Layer
        self.dropout = nn.Dropout(p=0.6)

        # TopK Pooling Layer for better feature selection
        self.pool = TopKPooling(hidden_channels * 4, ratio=0.8)  # Adjusted to match hidden dimension after GAT layers

    def forward(self, data):
        x, edge_index = data.x, data.edge_index

        # First GAT Layer with Residual Connection, ReLU, and Dropout
        x_1 = self.conv1(x, edge_index)
        x_1 = self.norm1(x_1)
        x_1 = F.relu(x_1)
        x_1 = self.dropout(x_1)

        # Second GAT Layer with Residual Connection, ReLU, and Dropout
        x_2 = self.conv2(x_1, edge_index)
        x_2 = self.norm2(x_2 + x_1)  # Residual Connection
        x_2 = F.relu(x_2)
        x_2 = self.dropout(x_2)

        # Third GAT Layer with Residual Connection, ReLU, and Dropout
        x_3 = self.conv3(x_2, edge_index)
        x_3 = self.norm3(x_3 + x_2)  # Residual Connection
        x_3 = F.relu(x_3)
        x_3 = self.dropout(x_3)

        # Apply TopKPooling for feature selection
        batch = torch.zeros(x_3.size(0), dtype=torch.long, device=x_3.device) if not hasattr(data, 'batch') else data.batch
        x_3, edge_index, _, batch, perm, _ = self.pool(x_3, edge_index, batch=batch)

        # Mask and label updates only for selected nodes, avoid modifying data directly
        data_out = data.clone()
        data_out.y = data_out.y[perm]
        data_out.train_mask = data_out.train_mask[perm]
        data_out.test_mask = data_out.test_mask[perm]

        # Final GAT Layer for Classification
        x = self.conv4(x_3, edge_index)

        return x, data_out
```

```python
# Model Initialization
num_node_features = data.num_node_features  # Feature count from the dataset
hidden_channels = 128  # Increase the hidden units for better representation learning
num_classes = dataset.num_classes  # Number of classes for classification

model = EnhancedGATWithAttention(num_node_features, hidden_channels, num_classes)
```

```python
# Step 3: Define Weighted Loss, Optimizer, and Scheduler
class_sample_count = torch.bincount(data.y[data.train_mask])
weight = 1. / class_sample_count.float()
weight = weight[:num_classes]  # Ensure weight tensor matches the number of classes
criterion = nn.CrossEntropyLoss(weight=weight)  # Weighted Loss

optimizer = optim.Adam(model.parameters(), lr=0.01, weight_decay=1e-3)  # Reduced learning rate and increased weight decay
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=50, gamma=0.5)
```

```python
# Step 4: Training Loop with Early Stopping
best_loss = float('inf')
patience = 20
trigger_times = 0

model.train()  # Set the model to training mode

for epoch in range(300):
    optimizer.zero_grad()  # Clear gradients
    out, data_out = model(data)  # Forward pass, using cloned data

    # Use the train_mask to get the relevant nodes for training
    out = F.log_softmax(out, dim=1)  # Apply log_softmax after getting node-wise predictions
    loss = criterion(out[data_out.train_mask], data_out.y[data_out.train_mask])  # Compute loss on training nodes
    loss.backward()  # Backpropagation
    optimizer.step()  # Update model parameters
    scheduler.step()  # Adjust learning rate

    # Early Stopping Check
    if loss.item() < best_loss:
        best_loss = loss.item()
        trigger_times = 0
    else:
        trigger_times += 1
        if trigger_times >= patience:
            print(f"Early stopping at epoch {epoch}")
            break

    if epoch % 10 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item()}')
```

```python
# Step 5: Evaluation
model.eval()  # Set the model to evaluation mode
out, data_out = model(data)
out = F.log_softmax(out, dim=1)  # Apply log_softmax for evaluation
_, pred = out.max(dim=1)  # Get predictions
correct = (pred[data_out.test_mask] == data_out.y[data_out.test_mask]).sum()  # Count correct predictions
accuracy = int(correct) / int(data_out.test_mask.sum())  # Calculate accuracy
print(f'Accuracy: {accuracy:.4f}')

# Step 6: Calculate F1 Score and Print Classification Report
f1 = f1_score(data_out.y[data_out.test_mask].cpu(), pred[data_out.test_mask].cpu(), average='weighted')
print(f'F1 Score: {f1:.4f}')

print(classification_report(data_out.y[data_out.test_mask].cpu(), pred[data_out.test_mask].cpu()))
```

```python
import torch
import torch.nn.functional as F
import torch.nn as nn
from torch_geometric.nn import GATConv, BatchNorm, LayerNorm, TopKPooling
from torch_geometric.data import Data
from sklearn.metrics import f1_score, classification_report
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import json
import time
```

```python
# Assume the trained GAT model is already defined as EnhancedGATWithAttention and loaded
# Trained Model Initialization
num_node_features = data.num_node_features
hidden_channels = 128
num_classes = dataset.num_classes
```

```python
model = EnhancedGATWithAttention(num_node_features, hidden_channels, num_classes).to('cuda')  # Move model to GPU if available
```

```python
model.load_state_dict(torch.load("enhanced_gat_model.pth"))  # Load the pre-trained model
model.eval()  # Set to evaluation mode
```

```python
def generate_node_embeddings(model, data):
    """
    Generate embeddings for each node using the trained GAT model.
    """
    data = data.to('cuda')  # Ensure data is on the same device as the model
    with torch.no_grad():
        out, _ = model(data)
        node_embeddings = out  # These are the node-wise embeddings from the GAT model
    return node_embeddings
```

```python
# Step 2: Prepare Inference Data
node_embeddings = generate_node_embeddings(model, data)

# Normalizing embeddings for better inference results (optional but recommended)
node_embeddings = F.normalize(node_embeddings, p=2, dim=1)
```

```python
# Load GPT-2 model and tokenizer once before the loop to avoid repeated loading
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
gpt2_model = GPT2LMHeadModel.from_pretrained("gpt2").to('cuda')  # Move model to GPU if available
```

```python
# Step 3: Define Inference Function Using GPT-2
def query_gpt2_inference(node_info, tokenizer, gpt2_model):
    """
    Use GPT-2 model to perform inference on a given node.
    """
    prompt = f"Node Information: {node_info}\nGenerate insights based on the given node information:"
    inputs = tokenizer.encode(prompt, return_tensors="pt", max_length=512, truncation=True).to('cuda')  # Move inputs to GPU

    # Generate output from GPT-2
    max_new_tokens = 50  # Set a smaller number of new tokens to generate
    output = gpt2_model.generate(inputs, max_new_tokens=max_new_tokens, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)
    decoded_output = tokenizer.decode(output[0], skip_special_tokens=True)
    return decoded_output
```

```python
# Step 4: Run Inference Over All Nodes
inference_results = []
num_nodes = node_embeddings.shape[0]

start_time = time.time()  # Start timing the inference loop

for node_idx in range(num_nodes):
    if node_idx >= node_embeddings.shape[0] or node_idx >= data.x.shape[0]:
        print(f"Skipping node {node_idx} due to out of range index")
        continue

    # Print progress every 100 nodes
    if node_idx % 100 == 0:
        elapsed_time = time.time() - start_time
        print(f"Processing node {node_idx}/{num_nodes} - Elapsed Time: {elapsed_time:.2f} seconds")

    node_vector = node_embeddings[node_idx].cpu().numpy().tolist()
    
    # Prepare node information to provide more context to the LLM
    node_info = {
        "embedding": node_vector,
        "text": data.x[node_idx].cpu().numpy().tolist(),  # Assuming node features contain text
        "neighbors": data.edge_index[:, data.edge_index[0] == node_idx].cpu().numpy().tolist() if node_idx < data.edge_index.shape[1] else []  # Getting neighbors
    }
    
    # Query the GPT-2 model for inference
    result = query_gpt2_inference(node_info, tokenizer, gpt2_model)
    if result:
        inference_results.append({"node_idx": node_idx, "inference": result})
```

```python
# Step 5: Analyze and Present Inference Results
# Here you could analyze the LLM outputs, store them in a file, or use them for further applications
for result in inference_results:
    node_idx = result["node_idx"]
    inference = result["inference"]

# Optional: Save results to file
with open("inference_results.json", "w") as f:
    json.dump(inference_results, f)
```
